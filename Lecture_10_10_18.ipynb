### Bernoulli Trial
Bernoulli is test which is true or false, yes or no, 1 or 0.<p>

$\begin{equation}
    f(k;p) =
    \left\{

                p & \mathrm{if\ } k=1 \\

        \end{array} 
    \right.
\end{equation}$

### Binomial Trials

Binomial Distribution = $\binom{n}{k}p^{n}(1-p)^{n-k}$

Where $\frac{n!}{k!(n-k)!} = \binom{n}{k}$

#### Example (by hand)
A Binomial test is a set of "n" bernoulli trials. For example, finding the probability that a 5 coin flips and all 5 end up all heads.<p>

Solution: We would plug in 5 for n, 5 for k, and .5 for p. $\binom{5}{5}.5^{5}(1-.5)^{5-5} = 1/32 = .03125 = 3.125  percent $

### Poisson Distribution

$P\left( k \right) = \frac{{e^{ - \lambda } \lambda ^k }}{{k!}}$

Where $\lambda$ is the mean.

#### Example (by hand) 
Suppose you knew that the mean number of calls to a fire station on a weekday is 8. What is the probability that on a given weekday there would be 11 calls? <p>
Solution: $P\left( 11 \right) = \frac{{e^{ - 8 } 8 ^{11} }}{{11!}} = .072 = 7.2 percent$

#### Example (by Python)

```python
from scipy.stats import poisson

poisson.pmf(k = 11, mu = 8) # Note that mu = mean is common notation in statistics
```

## Normal Distribution
Yes the formula looks ugly but don't get discouraged.

$ p(x) \sim \frac{1}{\sqrt{2\pi\sigma^2}} \exp{ \bigg[-\frac{1}{2}\bigg( \frac{x-\mu}{\sigma}\bigg)^2 \bigg] } $

$ X \sim \mathcal{N}(\mu,\,\sigma^{2})\,. $ This is common notation to say a random variable is normally distributed.

### Z-Score

$ z = \frac{x - \mu}{\sigma} $

#### Example (by hand)

Suppose you have a process that follows a standard normal distribution $ X \sim \mathcal{N}(0, 1)\,. $ What is the probability that we'll get a value between 1 and 2, $ P(1 <= X <= 2) $?

Solution: $ P(1 <= X <= 2) = P(X <= 2) - P(X <= 1).$ Look up Z-values on Z-table for the points 1 & 2. When I look up the z-table for 2 I find z = .9772 & .8413 for 1. Thus, .9772 - .8413 = .1359. This means I have a 13.59% chance of falling between 1 and 2 in a standard noraml process.

### Example (by Python)

```python
from scipy.stats import norm

z_at_2 = norm(0,1).cdf(2)
z_at_1 = norm(0,1).cdf(1)

probability = z_at_2 - z_at_1
probability
```

# Central Limit Theorem

The central limit theorem says that the distribution of the average of a SAMPLE from ANY RANDOM VARIABLE will be closely approximated by a normal distribution.

```python
from IPython.core.display import Image 
Image(filename='/Users/kevin/Dropbox/Github/Classes/IS_485_685/CLT.png')
```

### Example

Lets say we have randomly selected 100 bags that have an expected value (mean) of 50 pounds. If the variance weight of each bag is 1. Calculate $ P(49.8 <= \bar X <= 50.3) $<p>
Solution: Convert 50.3 and 49.8 to Z scores using CLT.<p>
(50.3 - 50 / (1 / $sqrt(100)$) ) = 3.<p>
(49.8 - 50 / (1 / $sqrt(100)$) ) = -2.<p>

Look up z @ 3 and -2 to get .9987 and .0228, thus .9987 - .0228 = .9759. Therefore, we have a 97.59percent chance of any bag weighing between 50.3 and 49.8 pounds.
