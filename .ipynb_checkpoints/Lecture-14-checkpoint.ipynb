### PCA (Principal Component Analysis)
The main objective of PCA is to decorrelate the data (reduce correlation among features) and the find the "intrinsic" value of a data set. The intrinsic value allows us to reduce the dimensionality of a data set, maybe 30 columns, and find the same amount of information in potentially 2 or 3 principal components (transformed columns). The main advantage of this is multi-fold as it now allows us to reduce multicollinearity (issues for models like linear regression), plus it reduces the amount of columns and values we need to store in order to run our model on.

### Example Data Set, Iris

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import tree

# Load Data
iris = datasets.load_iris()
print(iris.target_names)

# Make a dataframe for easier use later
X = iris.data
y = iris.target

# Make X a DataFrame
df_x = pd.DataFrame(X, columns=iris.feature_names)

# Make y a DataFrame
df_y = pd.DataFrame(y, columns=['Type_Flower'])

# Concatenate them together
df = pd.concat([df_x, df_y], axis = 1)
df.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Fancy 3D Plot
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-178, azim=40)

ax.scatter(df.iloc[:, 0], df.iloc[:, 1], df.iloc[:, 2], c=df.iloc[:,-1], cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("3D Representation of Sepal Width, Sepal Length, and Petal Length")
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')

plt.show()
plt.close()
fig.clf()
```

### Talking Points: How could we use PCA to reduce the 4 dimensions total (only 3 pictured here)?
Looking at the graph above we can see that there is in fact variance across all 3 dimensions, but that statement in itself is the beauty of PCA. We haven't done anything to this data, which is really 4 dimensions (columns), but we can already visually separate the data in 3 dimensions. That is what PCA will work to do with ANY number of columns and features. We want the most information "bang for the buck."

### PCA on all 4 dimensions
* Note: I only graphed 2 dimensions, sepal length versus sepal width in order for us to view where the first principal component is on the graph. The direction of the arrow is the first PCA feature.

```python
# Import PCA
from sklearn.decomposition import PCA

# Create PCA instance: model
model = PCA()

# Make a scatter plot of the untransformed points
plt.scatter(df.iloc[:,0], df.iloc[:,1])

# Create a PCA instance: model
model = PCA()

# Fit model to points
model.fit(df)

# Get the mean of the grain samples: mean
mean = model.mean_

# Get the first principal component: first_pc
first_pc = model.components_[0,:]

# Plot first_pc as an arrow, starting at mean
plt.arrow(mean[0], mean[1], first_pc[0]*3, # Multiply by 3 increases the length of the arrow
          first_pc[1]*3, 
          color='red', 
          width=0.02)

# Keep axes on same scale
plt.title('PCA Component 1')
plt.axis('equal')
plt.show()
```

### Talking points
Notice that the red line shows the largest variance, which is the most informative PCA feature. SKLearn naturally orders PCA features in terms of importance.

### PCA Variances, Unscaled

```python
# Perform the necessary imports
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

# Create scaler: scaler
scaler = StandardScaler()

# Create a PCA instance: pca
pca = PCA()

# Create pipeline: pipeline
pipeline = make_pipeline(pca)

# Fit the pipeline to 'samples'
pipeline.fit(df)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.title('PCA Variances Unscaled')
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```

### PCA Variances, Scaled

```python
# Perform the necessary imports
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

# Create scaler: scaler
scaler = StandardScaler()

# Create a PCA instance: pca
pca = PCA()

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,pca)

# Fit the pipeline to 'samples'
pipeline.fit(df)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.title('PCA Variances Unscaled')
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```

### Talking points
* Notice the difference between the scaled and unscaled versions of the pipeline within the model. The scaled version helps "pop" out the most important variances more so than the unscaled model. Again, scaling is almost always a good idea.
* Secondly, we can see that the number of PCA features drops off drastically around 2 features (maybe argue for 3), but ultimately this gives us our intrinsic value of the data set. We can say that the 4 dimensional data set can be explained by 2/3 PCA features and still not lose much information.

### TF-IDF Word Frequency Arrays
* Motivation: In natural language processing its a necessity to know which words are showing up in which documents & how often they arrive. Ultimately, we want the algorithm to learn what words are important and which aren't. Words like "Awesome" are much more informative than something common like "the" when doing sentiment analysis on documents (how does someone feel about something).
* We use TfidfVectorizer() from sklearn to convert to word frequencies, THEN we use a similar approach to PCA to extract the n most important features; n could be 5, 10, 100, whatever you decide, just like in standard PCA.
* What's different from PCA with natural language processing? Often in different documents we don't have the same informative words repeating themselves, thus we would have very "sparse" matrices. This means when we convert a word to a frequency for the algorithm to understand the word will mostly have zero as the value throughout most documents. We need a way to process this information more efficiently, so we use TruncatedSVD instead of PCA as it handles these matrices better (called CSR Matrices). TruncatedSVD is used in the same way as PCA other than that.

```python
# Example documents for easier visualization
# These documents could be thousands of words long. DataCamp has you do an exercise on Wikipedia articles.
documents = ['cats say meow', 'dogs say woof', 'dogs chase cats']
```

#### Create the tf-idf word frequency arrays.

```python
# Import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a TfidfVectorizer: tfidf
tfidf = TfidfVectorizer()

# Apply fit_transform to document: csr_mat
csr_mat = tfidf.fit_transform(documents)

# Print result of toarray() method
print(csr_mat.toarray())

# Get the words: words
words = tfidf.get_feature_names()

# Print words
print(words)
```

#### Pipeline & fit TruncatedSVD to pull the 3 most important words

```python
# Perform the necessary imports
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
import pandas as pd

# Create a TruncatedSVD instance: svd
svd = TruncatedSVD(n_components=3)

# Create a KMeans instance: kmeans
kmeans = KMeans(n_clusters=2)

# Create a pipeline: pipeline
pipeline = make_pipeline(svd,kmeans)

# Fit the pipeline to articles
pipeline.fit(csr_mat)

# Calculate the cluster labels: labels
labels = pipeline.predict(csr_mat)

print(labels)
```

### Talking points
Although this isn't an optimal example considering we only have 3 example documents of 3 words each, but it does show the walkthrough for doing TruncatedSVD() and pipelining with KMeans() in order to predict which documents belong with which through clustering. The algorithm decided that 'cats say meow' & 'dogs say woof' should be clustered together and 'dogs chase cats' should be by itself. This does make some intuitive sense.

### NMF (non-negative matrix factorization)
* Motivation: Why do we need ANOTHER dimension reduction technique? NMF has a few key benefits over PCA, as well as one major con.
* Major Pro: NMF models are interpretable, can take conventional and CSR Matrices as well.
* Major con: NMF only works on non-negative values, thus you couldn't use it if you had any negative values.
* You could use NMF for word frequencies, images encoded as arrays, audio spectrograms, purchase histories, or recommender systems.
